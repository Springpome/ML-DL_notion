# Decision Tree

Decision Tree는 다른 괜찮은 단어들을 생각해봤지만 스무고개 놀이가 제일 정확한거같다

예를들어 시험에서 A를 받은 데이터를 초록색 으로 표현했다고 하면 평소 공부시간이 얼마나 되는지 에서 Tree가 나뉠것이고 잠은 몇시간 자는지 에서 또 나뉠것이다 그 답에 따라서 Tree를 따라가다가 결국에는 이건 어떤 레이블이겠구나 라고 분류를 할수있는것이다

하지만 그만큼 질문을 잘 설계하는것이 중요할것이다 (쓰잘데기없는질문 넣어봤자 복잡해지기만 하고 제대로된 결과는 안나올테니)

Decision Tree의 몇가지 문제점이 있는데 그중 하나가 Over fitting이다 

Decision Tree랑 Over fitting이랑 뭔상관이냐 할수있겠지만
tree 구조가 훈련 데이터에 너무 의존적이라 현실의 데이터를 정확하게 나타내지 못하게 될수있다

Tree 구조가 커질수록 학습데이터에 알맞게 모델이 만들어지면 현실 데이터로 일반화(generaliztion)하기 어려워 질수 있다 


Decision Tree의 또다른 문제점은 생성하는 방식에서 한계가 있다는 것이다

Tree는 특정 상황에서 어떤 속성을 가지고 분할해야 정보 획득량이 가장클것이지 확인해서, 그 속성에따라 데이터를 분할한다.

그러다 보니 데이터를 분할하고 나서 나오는 데이터를 가지고 더나은 속성을 가지고 분할이 될수있음에도 즉,그 현재의 상황에서 만 생각하고 분할시키기 때문에 1차 분할이 안좋고 2차, 3차 분할이 좋은 상황은 보지 못하고 당장의 1차 분할이 좋은 상황만 가지고 분할시키기 때문에 더 좋은 Tree를 만들수 있음에도 그 가능성을 배제시켜버린다

이런 방법을 가지고 greedy 라고 부르는데 

여러 경우중 하나를 결정 해야 할때마다 그 순간에 최적이라고 생각되는것을 선택해나가는 방식으로 진행해서 최종적인 해답에 도달하는 '탐욕 알고리즘' 이 되는것이다.
