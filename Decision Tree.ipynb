## 1.Decision Tree의 전체적인 구조

Tree 구성 용어

image.png

* Root node : Work to do? 부분으로 Decision Tree가 시작되는 node
* Branch : edge라고도 부르며 node에서 node로 가는 길목
* Internal node : Decision node 라고도 부르며 하나의 branch로 가기 위해 '결정'을 해주는 node
* Leaf node : External node 또는 Terminal node라고도 부르며 분류문제에서 class(label)에 해당하는 node

머신러닝의 종류와 용어 개념에서 설명했던 Feature(Attribute)에 해당하는 부분은 Decision Tree 모형에서 Internal node에 해당한다


더 나아갈 Feature(Internal node)가 없으면 leaf node로 간주하고 자식 node를 생성하지 않는다.
또한 Leaf node가 더이상 없다면 가장 많은 수의 label이 있는것을 결과로 채택한다.


## 2.좋은 트리란?

Internal node는 '트리의 높이(깊이)'를 결정하는 요소이다
즉, Internal node의 개수가 증가할수록 트리는 높아진다(깊어진다)

좋은 트리란

"Leaf node의 한 쪽에 통일된 label 데이터들만 남으면서 트리의 높이가 짧은 트리"

이러한 조건을 충족시키기 위해서는 Feature의 선정 순서(Internal node 순서)가 중요하다.
적절한 Feature 순서를 선정하는 기준(수학적으로 잘계산하여 설정해야한다)은 데이터를 잘 나누어 줄 수 있는 Feature를 선정하는 것이다.

## 3.'ID3 알고리즘'

위에서 말한 적절한  Feature 순서를 선정하는 기준에서 수학적으로 잘 계산하라고 하였다.

그 수학적으로 계산을 하기 위해 사용하는 알고리즘이 'ID3 알고리즘' 이다.


image.png

* Purity : 한 쪽 데이터만 존재할 수록 more purity라고 할 수 있다. 위 그림상에서 Minimum impurity 에 해당하는 부분이 the best purity이다
* Entropy : Purity와 반대개념이다. Impurity라고도 하며 클술록 Chaotic(혼란스러움)하다. 따라서 Entropy 가 클수록 정보(Information)이 많음을 의미 한다.


Entropy를 구하는 수식

수식(값의 범위는 0~1)
$$\sum _i -p_ilog_2 p_i$$



* Information Gain(정보획득) : 자식 node들이 갖는 데이터가 pure할수록 좋다. 부모 노드의 Purity와 여러 자식 노드들의 Purity를 비교해서 차이가 가장 클수록 좋은 Feature이다

$$Information Gain(정보획득) = 부모노드의 Entropy - 자식노드의 Entropy$$

이때, 정보획득의 값은 0~1사이 이며, 자식 노드가 2개 이상이라면 정보획득을 구하는 공식에서 '자식노드의 Entropy 평균값'으로 대체한다

즉 **IG 가 큰 Feature를 선택하며 이것은 곧 데이터를 가장 잘 나누어 주는 Feature**라고 할수 있을것이다
(IG가 크다고 해서 무조건 Feature가 좋은 트리가 되진 않는다)

## 4.Tree생성 순서

* 노드 순서는 Root에서 Leaf 노드로 간다

1. 노드에서 고려할 데이터가 이미 하나의 class에 속하거나 더 이상 feature가 없어서 leaf node가 된경우, 가장 많은 수의 class를 결과값으로 채택
2. 각 노드에서 고려할 feature 생성시, IG가 가장 큰 feature를 선택(단, Leaf node에서 완벽하게 나누어진다는 보장은 없다)
3. 선택된 feature에 대한 조건별로 '자식 노드' 생성
4. 각 자식 노드에서 해당 조건을 만족하는 데이터끼리 1번 부터 반복

요약하자면 "처음에 feature별로 IG를 계산후 가장 높은 feature를 선정. 그리고 처음에 선정한 feature를 제외하고. 다음으로 IG가 높은 feature를 선정..." 이런식으로 프로세스가 진핸된다
