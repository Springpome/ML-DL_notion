## 1.Linear Regression(선형 회귀)

* 선형 모델은 입력 데이터에 대한 선형 함수를 만들어 예측 수행
* 회귀 분석을 위한 선형 모델은 다음과 같이 정의
$$ ŷ(w,x) = β_0 +β_1 x_1+...+β_p x_p$$
* $x$ : 입력 데이터
* $β$ : 모델이 학습할 파라미터
* $β_0$ : 편향
* $β_1∼β_p$ : 가중치

* 선형 회귀는 모델의 예측과 정답 사이의 평균제곱오차(Mean Squared Error) 를 최소화 하는 최적의 파라미터 β를 찾는것

* Cost Function : Loss Function의 일반화 버전으로서 모든 데이터와 실제 값들과의 차이의 종합이며 Loss의 합에 정규화 Term을 포함시키는 것이다.


Loss 값과 Cost 값을 구한후 Cost값을 최소화 함으로써 가장 적합한 파라미터를 찾을수있다

## 2.Ridge Regression

**Ridge Regression 은 Overfitting(과적합) 문제를 예방하기 위해 사용한다
즉, Linear Regrssion 에서 얻어지는 파라미터값이 너무 커지지 않도록 하는것이다.**

이 때 Regularization(규제) 를 이용해서 파라미터 값을 제한하게 되는데 Ridge는 L2 norm 방법을 사용한다.

**구체적으로 파라미터(베타)값을 0이아닌 0에 근사한 값으로 만듦 으로써 크기가 커지는 것을 제한한다.(모든 핏처가 주는 영향을 최소화(기울기를 작게 만듦))**

Regularization를 사용하면 다중공선성(multicollinearity) 문제를 방지하기 때문에 모델의 과대 적합(overfitting)을 막을 수 있게 됨

다중공선성 문제는 두특성이 일치에 가까울 정도로 관련성(상관관계)이 높을 경우 발생

* Linear regression
$$ ŷ(w,x) = β_0 +β_1 x_1+...+β_p x_p$$
$$\vecβ = argmin(y - \hat{y})^T(y-\hat{y})$$
* Ridge regression
$$\hat{β}^{Ridge} = argmin{(y-Xβ)^T(y-Xβ)+λβ^Tβ}$$

λ : 0이상의 값(주로 Cross - validation 등을 통해 결정)

수식을 간단하게 해석하자면 기존에 구했던 Linear Regression 에서 구한 파라미터값(베타값)을 Ridge로 적용했을때 기존의 베타값에다가 람다(λ)를 곱해준 베타값을 더해줌으로써 파라미터가 커지는 것을 제한하게 된다

## 3.Lasso Regression

Lasso 회귀는 기본적으로 Ridge 회귀와 목적은 동일하다. 하지만 파라미터를 어느정도까지 제한시키냐의 차이로 볼수있다. 즉, 파라미터값을 0에 근사하는 값으로 제한하는 Ridge 와 달리 Lasso는 람다값을 0으로 만들수 있음으로써 파라미터값을 제한하는 정도가 Ridge보다 크게 된다

Regularization(규제) 를 이용해서 파라미터 값을 제한하게 되는데 Lasso는 L1 norm 방법을 사용한다.

하지만 L1 norm 특성상 error를 계산하는 방식이 절대값을 이용하므로 미분이 '불가능'하다. 따라서 Cost function을 최소화 하기위한 방법으로 Gradient Descent(미분) 을 사용해야 하는데 이때 Lasso Regression은 막다른 길목에 다다르게 된다.

또한 Lasso는 Feature Selection효과를 가지고 있다. Feature Selection이란, 결과값 생성에 중요하지 않은 파라미터(가중치)값 이라면 0으로 만들어 버려 Feature를 선별(selection)하는것을 말한다.

* Lasso regression

$$ \hatβ^{Lasso} = argmin({(y-Xβ)^T(y-Xβ)+λ||β||_1}) $$

##4.Logistic Regression

Logistic Regression 은 Linear Classification 이라고도 하며 Linear(선형)으로 분류 문제를 해결할수있는 모델을 뜻한다.

* 로지스틱 회귀는 이름에 회귀라는 단어가 들어가지만, 가능한 클래스가 2개인 이진 분류를 위한 모델
* 로지스틱 회귀의 예측 함수 정의
$$ σ(x)= {1\over1+e^{-x}} $$
$$ŷ=σ(w_0+w_1x_1+...+w_px_p)$$
* 로지스틱 회귀 모델은 선형 회귀 모델에 시그모이드(σ) 함수를 적용
* 로지스틱 회귀의 목표는 다음과 같은 목적 함수를 최소화 하는 파라미터 w를 찾는것
$$BinaryCrossEntropy=-{1\over N}\sum_{i=1}^Ny_i\log(ŷ_i)+(1-y_i)\log(1-ŷ_i)$$



### 4-1확률적 경사 하강법(Stochastic Gradient Descent)



* 학습 파라미터에 대한 손실 함수의 기울기를 구해 기울기가 최소화 되는 방향으로 학습
$${∂L \over ∂w} = \lim_{h→0}{L(w+h)-L(w) \over h}$$
$$ w' = w-α{∂L\over ∂w} $$


위와 같이 Logistc Regression을 통해서 Linear로 분류 문제를 해결할수있다.
